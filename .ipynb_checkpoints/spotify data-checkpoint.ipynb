```python
import zipfile
import pandas as pd
import datetime

# List to hold the data frames
dfs = []

# Open the ZIP file containing your Spotify JSON files
with zipfile.ZipFile('my_spotify_data.zip', 'r') as zf:
    for name in zf.namelist():
        # Process only audio streaming history files
        if 'Streaming_History_Audio_' in name:
            print("Processing:", name)
            df_temp = pd.read_json(zf.open(name))
            dfs.append(df_temp)

# Combine all dataframes into one
df = pd.concat(dfs, ignore_index=True)

# Convert timestamp string into datetime objects
df['ts'] = pd.to_datetime(df['ts'])

# Quick summary of the data
print("Columns:", df.columns.tolist())
print("Number of streaming events:", df.shape[0])
total_time = datetime.timedelta(milliseconds=int(df['ms_played'].sum()))
print("Total streaming time:", total_time)
```

```python
# Group by artist name and sum the listening time (ms_played)
top_artists = df.groupby('master_metadata_album_artist_name')['ms_played'].sum().sort_values(ascending=False)

# Convert from milliseconds to hours
top_artists_hours = top_artists / 3.6e+6

# Show top 60 artists
print("Top 60 Artists (by listening hours):")
print(top_artists_hours.head(60))
```

```python
import matplotlib.pyplot as plt

# Select the top 10 artists for plotting
top10_artists = top_artists_hours.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10_artists.index, top10_artists.values, color='skyblue')
plt.xlabel("Listening Time (Hours)")
plt.title("Top 10 Artists by Listening Time")
plt.gca().invert_yaxis()  # so the top artist is on top
plt.show()
```

```python
# Filter your data by the year 2015
df_2015 = df[df['ts'].dt.year == 2023].copy()

# Group by artist and track name, and sum listening time
df_tracks_2015 = df_2015.groupby(['master_metadata_album_artist_name', 'master_metadata_track_name', 'spotify_track_uri'])['ms_played'].sum()

# Sort to get the most played songs
df_tracks_2015 = df_tracks_2015.sort_values(ascending=False)

# Get the top 10
top10_tracks_2015 = df_tracks_2015.head(10)

# Create a DataFrame to show artist, song, and uri neatly
top10_tracks_2015_df = pd.DataFrame(top10_tracks_2015.index.tolist(), 
                                    columns=['Artist', 'Song', 'Spotify URI'])
print("Top 10 Songs of 2015:")
print(top10_tracks_2015_df)
```

```python
# Identify the top 5 artists overall
top5_artists = top_artists.head(5).index.tolist()

plt.figure(figsize=(10, 6))
for artist in top5_artists:
    artist_df = df[df['master_metadata_album_artist_name'] == artist][['ts', 'ms_played']].copy()
    # Sort by time
    artist_df = artist_df.sort_values('ts')
    # Compute a rolling sum over 365 days; note: this requires a datetime index
    artist_df.set_index('ts', inplace=True)
    rolling_time = artist_df['ms_played'].rolling('365D').sum() / 3.6e+6  # convert ms to hours
    plt.plot(rolling_time.index, rolling_time, label=artist)

plt.xlabel("Time")
plt.ylabel("Listening Time (Hours, rolling 365 days)")
plt.title("Top 5 Artists Over Time")
plt.legend()
plt.show()
```
